You need the find the Kth most frequent alphabet in the string. If more than one alphabet share the Kth frequency, you need to find the lexicographically smallest alphabet among them. In case no Kth most popular alphabet exists, you need to write "NONE" (without the inverted commas).

Input Format

The first line contains T, the number of test cases in the input.

This is follwed by T test cases, each test case containing 2 lines.

The first line of each test case contains the values N and K for the test case, and the second line contains the string.

Constraints

1 <= T <= 100
1 <= N <= 100000
1 <= K <= N

```python
def find_index_freq_element(sorted_freq, k):
    res_stack = []
    #all chars with the same amount will be ignored in the stack 
    res_stack.append(sorted_freq[0])
    for i in range(1, len(sorted_freq)):
        if res_stack[-1][1] != sorted_freq[i][1]:
            res_stack.append(sorted_freq[i])             
    if len(res_stack) < k:
        return "NONE"
    return res_stack[k-1][0]                              
    
    

def kth_most_frequent(string, k):    
            
    # Step 1: Calculate frequencies
    freq = {}
    for char in string:        
        freq[char] = freq.get(char,0)+1
        
    # Step 2: Sort characters by frequency and lexicographically
    sorted_freq = sorted(freq.items(), key=lambda item: (-item[1], item[0]))
    #print("frequency and lexicographically = ", sorted_freq)
        
    # Step 3: Identify the Kth most frequent character       
    # (6 3) -> return c; if several K return first lexicographically sorted  
    # aaabbc
    #aaa - 1 freq  
    #bb - 2 freq 
    #c - 3 freq  
        
    #8 3
    #aaabbbcc
    # both 'a' and 'b' have the highest frequency of 3, and 'c' has the second highest frequency of 
    # Since there exists no alphabet with highest frequency, the answer is 'NONE'.
                            
    #Calculate freq
    return find_index_freq_element(sorted_freq, k)
     



#test
#Sample Output 0
#1. c
#2. a
#3. NONE
#4. b
# Sample input
test_cases = [
    ((6, 3), "aaabbc"),
    ((5, 1), "aaabb"),
    ((8, 3), "aaabbbcc"),
    ((5, 2), "aabcd")
]

# Get results
results = []
for tc in test_cases:
    string = tc[1]
    k = tc[0][1]
    results.append(kth_most_frequent(string, k))

# Print results
for result in results:
    print(result)
```

```python
# function to sort elements by frequency
def sort_freq(s, k):
    # 1. Store frequencies of characters (char -> count)
    m = {}
    for char in s:        
        m[char] = m.get(char,0)+1
    v=list(m.items())
 
    # Sort the element of array by frequency
    v.sort(key=lambda x:x[1],reverse= True)
 
    # Find k-th most frequent item. 
    count = 0
    for i in range(len(v)): 
        # Increment not same
        if (i == 0 or v[i][1] != v[i - 1][1]):
            count+=1
        if (count == k):
            return v[i][0]
    return "NONE"   
 
#On hackerRanks you have to filter string - "\n\r..."
#filtered_string = ''.join([char for char in string if char.islower()])

print(sort_freq("aaabbc", 3))
print(sort_freq("aaabb", 1))
print(sort_freq("aaabbbcc", 3))
print(sort_freq("aabcd", 2))
```

### Точка і відрізки

Дано n відрізків та m точок на числовій прямій. Для кожної з точок визначте, скільком відрізкам вона належить. Точка x належить відрізку з кінцями в a і b, якщо виконується нерівність min(a, b) <= x <= max(a, b).

Рішення має працювати швидше за О(n*m).

Input Format

Перший рядок містить два цілих числа n – число відрізків та m – число точок. У наступних n рядках по два цілі числа ai і bi - координати кінців відповідного відрізка. В останньому рядку m цілих чисел - координати точок.

Constraints

1 <= n <= 10^5
1 <= m <= 10^5
Output Format

Виведіть m чисел – для кожної точки кількість відрізків яким вона належить.

Sample Input 0

3 2
0 5
-3 2
7 10
1 6
Sample Output 0

2 0

```python
def track_users_online_L_R(segments, dots):
    events = []
    point_indices = {}
    
    # Create events for segment starts and ends
    for start, end in segments:
        events.append((start, 'L'))
        events.append((end, 'R'))
    
    # Create events for points
    for i, dot in enumerate(dots):
        events.append((dot, 'P'))
        if dot not in point_indices:
            point_indices[dot] = []
        point_indices[dot].append(i)
    
    # Sort events: segment start (L), point (P), segment end (R)
    events.sort(key=lambda x: (x[0], x[1]))

    active_segments = 0
    result = [0] * len(dots)
    
    # Sweep through events
    for position, typ in events:
        if typ == 'L':
            active_segments += 1
        elif typ == 'R':
            active_segments -= 1
        else:  # typ == 'P'
            for idx in point_indices[position]:
                result[idx] = active_segments
    
    return result



def track_users_online(segments, dots):
    #1. find min = 'start segment' and max = 'end segment' 
    start_event = min(segments, key=lambda x: x[0])[0]
    stop_event = max(segments, key=lambda x: x[1])[1]
    #2. sort all events 
    
    start_dict = {}
    stop_dict = {}
    dict_dots = {dot: None for dot in dots}
    #3. track 'start' and 'stop' event points 
    for start, stop in segments:
        start_dict[start] = start_dict.get(start, 0)+1
        stop_dict[stop] = stop_dict.get(stop, 0)+1
        
    #4. go through timeline for min to max    
    res_users_online = []
    counter_online = 0 
    for i in range(start_event, stop_event+1):
        #5. track counter "event 'start' +=1; event 'stop' -=1"
        counter_online+=start_dict.get(i,0)
        counter_online-=stop_dict.get(i,0)
        if i in dict_dots:
            res_users_online.append(counter_online)
            
    return res_users_online

segments = [[0, 5], [-3, 2], [7, 10]]
dots = [1,6]
#print(segments)
#print(dots)
res = track_users_online_L_R(segments, dots)
print(" ".join(str(c) for c in res))
```

# 3 sum closest

https://leetcode.com/problems/3sum-closest/


Given an integer array of length n and an integertarget, find 3 integers in the array such that the sum is closest to target.

Input Format

First line contains an array values, second line contains target value.

Constraints

3 <= n <= 100000
Output Format

Return the sum of the 3 integers. You may assume that each input would have exactly one solution.

Sample Input 0

-1 2 1 -4
1
Sample Output 0

2
Sample Input 1

0 0 0
1
Sample Output 1

0




```python
# 1. sort array
# 2. two pointers L and R. if sum < target => L++ else R--
# 3. Finsd the first event when closest sum become 'less or more'

# [1,2,10,12]
#  L-------R
from typing import List

class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        if len(nums) < 3:
            return -1
        
        nums = sorted(nums)
        closest_sum = sum(nums[:3])
        diff = float("inf")
        
        l, m, r = 0, 1, len(nums) - 1
        
        while l < len(nums) - 2:
            current_sum = nums[l] + nums[m] + nums[r]
            current_diff = abs(current_sum - target)
            
            if current_diff < diff:
                diff = current_diff
                closest_sum = current_sum
            
            if current_sum < target:
                m += 1
            elif current_sum > target:
                r -= 1
            else:
                return current_sum
            
            if m >= r:  # Reset pointers
                l += 1
                m = l + 1
                r = len(nums) - 1
        
        return closest_sum


# Input: nums = [-1,2,1,-4], target = 1
# Output: 2    
Solution().threeSumClosest([-1,2,1,-4], 1)         
```

Problem
Submissions
There are N classes in a school, each with M students. There is going to be a race of 100m dash, and a representative from each class will be chosen to participate in this race. You were assigned a task to choose these representatives. Since you did not want the race to be one sided, you wanted to choose the representatives such that the difference between the ability of the best representative and the worst representative is minimal.

For example,

if N = 3 and M = 4, and each class has students with following abilities:

Class 1: {12, 16, 67, 43}

Class 2: {7, 17, 68, 48}

Class 3: {14, 15, 77, 54}

it is best to choose the student with ability 16 from Class 1, 17 from Class 2, and 15 from Class 3. Thus, the difference in this case would be 17 - 15 = 2.

Your task is to calculate the minimal possible difference you can achieve by choosing a representative from each class.

Input Format

The first line of the input consists of two integers, N and M

The next N lines will have M integers. The jth element of ith line is the ability of the jth student in ith class.

Constraints

1 <= N <= 1000

1 <= M <= 1000

The numbers are between 0 and 10^9, inclusive.

Output Format

Output the minimal difference one can achieve by choosing the representative from each class.

Sample Input 0

3 4
12 16 67 43
7 17 68 48
14 15 77 54
Sample Output 0

2

```python
from collections import defaultdict

def flatten_and_sort(arrays):
    all_elements = []
    for i, arr in enumerate(arrays):
        for num in arr:
            all_elements.append((num, i))
    all_elements.sort()
    return all_elements

def find_smallest_range(arrays):
    all_elements = flatten_and_sort(arrays)
    left, right = 0, 0
    freq_counter = defaultdict(int)
    unique_arrays = 0
    min_diff = float('inf')
    start, end = -1, -1
    total_arrays = len(arrays)
    
    while right < len(all_elements):
        num, array_index = all_elements[right]
        freq_counter[array_index] += 1
        if freq_counter[array_index] == 1:
            unique_arrays += 1
        
        while unique_arrays == total_arrays:
            current_start, start_array_index = all_elements[left]
            current_end, end_array_index = all_elements[right]
            current_diff = current_end - current_start
            if current_diff < min_diff:
                min_diff = current_diff
                start, end = current_start, current_end
            
            freq_counter[start_array_index] -= 1
            if freq_counter[start_array_index] == 0:
                unique_arrays -= 1
            left += 1
        
        right += 1
    
    return min_diff, start, end


# Priority Queue (Min-Heap) Approach
import heapq

def initialize_heap(arrays):
    min_heap = []
    max_value = float('-inf')
    
    # Insert the first element from each array into the heap
    for i, arr in enumerate(arrays):
        heapq.heappush(min_heap, (arr[0], i, 0))
        max_value = max(max_value, arr[0])
    
    return min_heap, max_value

def update_range(min_heap, max_value, arrays):
    min_diff = float('inf')
    start, end = -1, -1
    
    while min_heap:
        min_value, array_index, element_index = heapq.heappop(min_heap)
        
        # Update the minimum difference
        current_diff = max_value - min_value
        if current_diff < min_diff:
            min_diff = current_diff
            start, end = min_value, max_value
        
        # Move to the next element in the array
        if element_index + 1 < len(arrays[array_index]):
            next_element = arrays[array_index][element_index + 1]
            heapq.heappush(min_heap, (next_element, array_index, element_index + 1))
            max_value = max(max_value, next_element)
        else:
            # If we run out of elements in any array, we cannot proceed further
            break
    
    return min_diff, start, end

def find_smallest_range_mean_heap(arrays):
    # Initialize the min-heap and max_value
    min_heap, max_value = initialize_heap(arrays)
    
    # Update and find the smallest range
    return update_range(min_heap, max_value, arrays)

# Example usage
arrays = [
    [12, 16, 67, 43],
    [7, 17, 68, 48],
    [14, 15, 77, 54]
]

#min_diff, start, end = find_smallest_range(arrays)
min_diff, start, end = find_smallest_range_mean_heap(arrays)
print(f"Minimum Difference: {min_diff}, Start: {start}, End: {end}")  # Output: Minimum Difference: 2, Start: 15, End: 17
```
