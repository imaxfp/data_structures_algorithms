### 2. Add Two Numbers

https://leetcode.com/problems/add-two-numbers/description

```python
# Definition for singly-linked list.
from typing import Optional

class ListNode:    
    def __init__(self, val=0, next=None):        
        self.val = val
        self.next = next
        
class Solution:    
    def __init__(self):
            self.res_head = None
            self.res_tmp = None
            
    def process_reminder_lists(self, l, carry):
        while l:
            sum_unit = l.val + carry
            carry = sum_unit // 10
            stor_d = sum_unit % 10
            self.res_tmp.next = ListNode(stor_d)
            self.res_tmp = self.res_tmp.next
            l = l.next 
        return carry                        
            
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:                
        #[9,9,9,9,9,9,9]
        #[9,9,9,9]
        # 8 9 9 9 0 0 0 1

        #1. Go via lists simultaneously while l1 or l2
        #2. add numbers and hold reminder for the next iteration
        #3. do simple math 
        #4. res to new ListNodes
                                
        
        carry = 0
        while l1 and l2:
            sum_unit = l1.val + l2.val + carry
            carry = sum_unit // 10
            stored_digit = sum_unit % 10
            
            if not self.res_tmp:
                self.res_tmp = ListNode(stored_digit)
                self.res_head = self.res_tmp
            else:
                self.res_tmp.next = ListNode(stored_digit)
                self.res_tmp = self.res_tmp.next
            l1 = l1.next
            l2 = l2.next   
        
        #process reminder of lists
        carry = self.process_reminder_lists(l1, carry)
        carry = self.process_reminder_lists(l2, carry)   
        
        # add ccarry reminder 
        if carry > 0:
            self.res_tmp.next = ListNode(carry)
                                                         
        return self.res_head
    
    
#l1 = [9,9,1]
#l2 = [1] 
#Expected = [0,0,2]
l1 = ListNode(9)
l1.next = ListNode(9)
l1.next.next = ListNode(1)
l2 = ListNode(1)
r = Solution().addTwoNumbers(l1, l2)
while r:
    print(r.val)
    r = r.next
 
#[9,9,9,9,9,9,9]
#[9,9,9,9]
#Expected [8,9,9,9,0,0,0,1]
l1 = ListNode(9)
l1.next = ListNode(9)
l1.next.next = ListNode(9)
l1.next.next.next = ListNode(9)
l1.next.next.next.next = ListNode(9)
l1.next.next.next.next.next = ListNode(9)
l1.next.next.next.next.next.next = ListNode(9)

l2 = ListNode(9)
l2.next = ListNode(9)
l2.next.next = ListNode(9)
l2.next.next.next = ListNode(9)
#r = Solution().addTwoNumbers(l1, l2)


#Input: l1 = [2,4,3], l2 = [5,6,4]
#Output: [7,0,8]
l1 = ListNode(2)
l1.next = ListNode(4)
l1.next.next = ListNode(3)

l2 = ListNode(5)
l2.next = ListNode(6)
l2.next.next = ListNode(4)            
```
